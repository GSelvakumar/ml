```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import plotly.express as px
import plotly.offline as py

%matplotlib inline 

data = pd.read_csv("/content/drive/MyDrive/Machine_Learning/COVID 19 Vaccine analysis/country_vaccinations.csv")
data.head()
```

##Data analysis using Covid 19 Vaccine data

```python
data.country.value_counts()
```

```python
data = data[data.country.apply(lambda x: x not in ["England", "Scotland", "Wales", "Northern Ireland"])]
data.country.value_counts()
```

```python
data.vaccines.value_counts()
```

```python
df = data[["country", "vaccines"]]
df[0: 10]
```

```python
dict = {}

for i in df.vaccines.unique():
   dict[i] = [df["country"][j] for j in df[df["vaccines"]==i].index]

vaccines = {}

for key, value in dict.items():
  vaccines[key] = set(value)

for i, j in vaccines.items():
  print(f"{i}:>>{j}")
```

```python
vaccine_map = px.choropleth(data, locations = 'iso_code', color= 'vaccines')
vaccine_map.update_layout(height=300, margin={"r":0, "t":0, "l":0, "b":0})

vaccine_map.show()
```
