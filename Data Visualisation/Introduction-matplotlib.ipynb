## Basic ploting with plot labels

**legend( )** is a method used to show the labels of the plot in the graph area.

```python
x = np.linspace(0, 2, 100)
plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')
plt.xlabel('xlabel')
plt.ylabel('ylabel')
plt.title('3 quadrants')
plt.legend()
plt.show()
```

```python
#Line Plot

lp_x = [2, 3, 4, 5]
lp_y = [9, 2, 1, 3]
plt.plot(lp_x, lp_y, marker='o', color='r', linestyle='--')
plt.show()
```

```python
#Bar Plot

bp_x = [2, 3, 4, 5]
bp_y = [9, 2, 1, 3]
plt.bar(bp_x, bp_y)
plt.show()
```

```python
#Scatter Plot

sp_x = [2, 3, 4, 5]
sp_y = [9, 2, 1, 3]
plt.scatter(sp_x, sp_y, marker='o', color='r')
plt.show()
```
