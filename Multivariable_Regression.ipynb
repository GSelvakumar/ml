## Data exploration with Pandas dataframe



```python
# Create a pandas dataframe
data = pd.DataFrame(data=boston_dataset.data, columns=boston_dataset.feature_names)

# Add price (target) column to the dataframe 
data['PRICE'] = boston_dataset.target
```

## Correlation

## $$ \rho _{XY} = corr(X, Y) $$

## $$  -1.0 \leq \rho _{XY} \leq +1.0 $$

```python
data["PRICE"].corr(data["RM"])
```

```python
data.corr()
```

```python
mask = np.zeros_like(data.corr())
triangle_indices = np.triu_indices_from(mask)
mask[triangle_indices] = True
mask
```

```python
plt.figure(figsize=(16, 10))
sns.heatmap(data.corr(), mask=mask, annot=True, annot_kws={"size": 14})
sns.set_style('white')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```

```python
nox_dis_corr = round(data['NOX'].corr(data['DIS']), 3)

plt.figure(figsize=(9, 6))
plt.scatter(x=data["DIS"], y=data["NOX"], alpha=0.6, s=80, color='indigo')
plt.title(f"DIS vs NOX (correlation is {nox_dis_corr})", fontsize=14)
plt.xlabel("Distance from employment", fontsize=14)
plt.ylabel("Nitric Oxide Pollution", fontsize=14)
plt.show()
```

```python
sns.set()
sns.set_context("talk")
sns.set_style("whitegrid")
sns.jointplot(x=data["DIS"], y=data["NOX"], height=7, color="blue", kind='hex')
plt.show()
```

```python
sns.set()
sns.set_context("talk")
sns.set_style("whitegrid")
sns.jointplot(x=data["TAX"], y=data["RAD"], height=7, color="darkred", joint_kws={"alpha":0.7})
plt.show()
```

```python
sns.lmplot(data=data,x='TAX', y='RAD', height=7)
```

```python
rm_tgt_corr = round(data['RM'].corr(data['PRICE']), 3)

plt.figure(figsize=(9, 6))
plt.scatter(x=data["RM"], y=data["PRICE"], alpha=0.6, s=80, color='green')
plt.title(f"RM vs PRICE (correlation is {rm_tgt_corr})", fontsize=14)
plt.xlabel("RM - Median Nr of Rooms", fontsize=14)
plt.ylabel("Price - Property price in 000s", fontsize=14)
plt.show()
```

```python
sns.lmplot(data=data, x='RM', y='PRICE', height=7)
```
